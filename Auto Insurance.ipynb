```python
import pandas as pd
```

```python
data=pd.read_csv("Insurance_Marketing-Customer-Value-Analysis.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
state=pd.get_dummies(data["State"], drop_first=True)

response=pd.get_dummies(data["Response"], drop_first=True)

coverage=pd.get_dummies(data["Coverage"], drop_first=True)

education=pd.get_dummies(data["Education"], drop_first=True)

employmentStatus=pd.get_dummies(data["EmploymentStatus"], drop_first=True)

renewOfferType=pd.get_dummies(data["Renew Offer Type"], drop_first=True)

```

```python
policyType=pd.get_dummies(data["Policy Type"], drop_first=True)

salesChannel=pd.get_dummies(data["Sales Channel"], drop_first=True)

vehicleClass=pd.get_dummies(data["Vehicle Class"], drop_first=True)

vehicleSize=pd.get_dummies(data["Vehicle Size"], drop_first=True)
```

```python
data_1=pd.concat([data,state, response, coverage, education, employmentStatus, renewOfferType, policyType, salesChannel, vehicleClass, vehicleSize], axis=1)
```

```python
data_1.drop(['State','Response', 'Coverage', 'Education', 'EmploymentStatus', 'Renew Offer Type', 'Policy Type', 'Sales Channel', 'Vehicle Class', 'Vehicle Size','Customer'], axis=1, inplace=True)
```
